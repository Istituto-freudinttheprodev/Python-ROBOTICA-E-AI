```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#Genera dati di esempio
data = {
    'Variable n°1': [1, 2, 3, 4, 5],
    'Variable n°2': [1, 2, np.nan, 4, np.nan],
    'Colonne mancanti': ['A', 'B', 'A', 'C', np.nan]
}
df = pd.DataFrame(data)
df1 = pd.DataFrame()
#Seleziona le colonne numeriche e riempi i valori NaN con la media
colonnenumeriche = df.select_dtypes(include=['number'])
df1[colonnenumeriche.columns] = colonnenumeriche.fillna(colonnenumeriche.mean())
#Seleziona le colonne categoriche e riempi i valori NaN con la moda
colonnecategoriche = df.select_dtypes(exclude=['number'])
df1[colonnecategoriche.columns] = colonnecategoriche.fillna(colonnecategoriche.mode().iloc[0])
# Creazione di grafici a barre per ogni variabile numerica
plt.figure(figsize=(14, 6))
# Grafico per la Variable n°1
plt.subplot(1, 2, 1)
sns.barplot(x=['Originali', 'Riempiti'], y=[df['Variable n°1'].mean(), df1['Variable n°1'].mean()])
plt.title('Variable n°1: confronto tra gli originali e i riempiti')
plt.ylabel('Media dei valori')
# Grafico per Variable n°2
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 2)
sns.barplot(x=['Originali', 'Riempiti'], y=[df['Variable n°2'].mean(), df1['Variable n°2'].mean()])
plt.title('Variable n°2: confronto tra gli originali e i riempiti')
plt.ylabel('Media dei valori')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
#Dati di esempio
data = {
    'Variable1': [1, 2, 3, 4, 5],
    'Variable2': [1, 2, np.nan, 4, np.nan],
    'Missing Column': ['A', 'B', 'A', 'C', np.nan]
}
#Creare un DataFrame
df = pd.DataFrame(data)
df1=pd.DataFrame()
def gestionemissingvalues(df):#questa è una funzione che si occupa solo dei missing values, in modo da poterla riusare in ogni momento
    #Trattamento dei missing values nelle variabili numeriche e categoriche
    colonnenumeriche=df.select_dtypes(include=['number'])
    colonnecategoriche=df.select_dtypes(exclude=['number'])
    df1[colonnenumeriche.columns] = df[colonnenumeriche.columns].fillna(df[colonnenumeriche.columns].mean())    
    df1[colonnecategoriche.columns] = df[colonnecategoriche.columns].fillna(df[colonnecategoriche.columns].mode().iloc[0])
    return df1
def main():
    df1=gestionemissingvalues(df)
    print(f"Questo è il primo con i valori mancanti: \n{df}\nQuesto invece è il secondo con i missing values sostituiti:\n{df1}\n")
if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
#Dati puramente casuali a solo scopo scientifico
np.random.seed(42)#il seme (il seed) serve per rendere i risultati casuali riproducibili ogni volta, se cambiassi il numero 42 cambierei il seme e cambiarei quindi anche la casualità del seme stesso
data={
    'Età': np.random.randint(18, 70, size=1000),
    'Genere': np.random.choice(['Maschio', 'Femmina'], size=1000),
    'Punteggio': np.random.uniform(0, 100, size=1000),
    'Reddito': np.random.normal(50000, 15000, size=1000)
}
df=pd.DataFrame(data)
#Istogramma dell'età
plt.figure(figsize=(10, 6))
sns.histplot(df['Età'], kde=True)#kde impostato su True mostra le barre più chiare, mentre su False più scure, si tratta soltanto di un effeto grafico
plt.title('Distribuzione dell\'età')
plt.xlabel('Età')
plt.ylabel('Frequenza')
plt.show()
#Grafico a barre del genere
plt.figure(figsize=(10, 6))
sns.countplot(x='Genere', data=df)
plt.title('Conteggio per genere')
plt.xlabel('Genere')
plt.ylabel('Conteggio')
plt.show()
#Scatter plot dell'età vs punteggio
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Età', y='Punteggio', data=df)
plt.title("Relazione tra l'età e il punteggio")
plt.xlabel('Età')
plt.ylabel('Punteggio')
plt.show()
#Box plot del reddito per genere
plt.figure(figsize=(10, 6))
sns.boxplot(x='Genere', y='Reddito', data=df)
plt.title('Distribuzione del reddito per genere')
plt.xlabel('Genere')
plt.ylabel('Reddito')
plt.show()
#Violin plot del reddito per genere
plt.figure(figsize=(10, 6))
sns.violinplot(x='Genere', y='Reddito', data=df)
plt.title('Distribuzione del reddito per genere')
plt.xlabel('Genere')
plt.ylabel('Reddito')
plt.show()
#Plotly Scatter Plot dell'età, punteggio e reddito
fig = px.scatter(df, x='Età', y='Punteggio', color='Reddito', hover_data=['Genere'])
fig.update_layout(title="Relazione tra l'età, il punteggio e il reddito", xaxis_title='Età', yaxis_title='Punteggio')
fig.show()
```

```python
# Importa le librerie necessarie.
import numpy as np          #libreria per operazioni numeriche e matrici.
import seaborn as sns       #libreria per la visualizzazione avanzata dei dati.
import matplotlib.pyplot as plt  #libreria per la creazione di grafici.
#Genera un dataset di esempio con variabili numeriche.
np.random.seed(42)#imposta il seed per la generazione di numeri casuali, garantendo la riproducibilità.
data = pd.DataFrame(np.random.rand(100, 5), 
                    columns=['Variabile1', 'Variabile2', 'Variabile3', 'Variabile4', 'Variabile5'])  # Crea un DataFrame con 100 righe e 5 colonne di valori casuali tra 0 e 1.
#Aggiungere alcune variabili categoriche generate casualmente.
data['Categoria1'] = np.random.choice(['A', 'B', 'C'], size=100)  # Aggiunge una nuova colonna 'Categoria1', con valori casuali tra 'A', 'B' e 'C'.
data['Categoria2'] = np.random.choice(['X', 'Y'], size=100)  # Aggiunge una nuova colonna 'Categoria2', con valori casuali tra 'X' e 'Y'.
# Calcola la matrice di correlazione delle variabili numeriche nel DataFrame.
correlation_matrix = data.corr()  # Utilizza il metodo 'corr()' per calcolare la correlazione tra le colonne numeriche.
# Imposta la dimensione della figura per il grafico. 
# Il parametro figsize=(larghezza, altezza) definisce le dimensioni della figura in pollici.
plt.figure(figsize=(10, 8))

# Crea un heatmap utilizzando Seaborn. 
# 'correlation_matrix' è la matrice di correlazione che vuoi visualizzare.
sns.heatmap(correlation_matrix, 
            annot=True,  # 'annot=True' aggiunge le etichette (i valori di correlazione) in ciascuna cella del heatmap.
            cmap='coolwarm',  # 'cmap' definisce la palette di colori da usare. 'coolwarm' è una palette da blu (valori bassi) a rosso (valori alti).
            fmt='.2f',  # 'fmt' specifica il formato numerico delle etichette. Qui si usano due cifre decimali.
            alpha=0.7)  # 'alpha' definisce la trasparenza delle celle. 0 è completamente trasparente, 1 è completamente opaco.

# Imposta il titolo del grafico.
plt.title("Matrice di correlazione")

# Mostra il grafico. Questa funzione rende visibile il grafico a schermo.
plt.show()
```

```python
import pandas as pd
import numpy as np

# Imposta il seed per la generazione di numeri casuali per rendere i risultati riproducibili.
np.random.seed(41)

# Crea un DataFrame vuoto.
df = pd.DataFrame()

# Genera dati casuali per il DataFrame.
n_rows = 100  # Numero di righe nel DataFrame.

# Genera una colonna di dati categorici casuali scegliendo tra 'A', 'B', e 'C'.
df['CatCol1'] = np.random.choice(['A', 'B', 'C'], size=n_rows)

# Genera un'altra colonna di dati categorici casuali scegliendo tra 'X' e 'Y'.
df['CatCol2'] = np.random.choice(['X', 'Y'], size=n_rows)

# Genera una colonna di dati numerici da una distribuzione normale standard (media 0, deviazione standard 1).
df['NumCol1'] = np.random.randn(n_rows)

# Genera una colonna di dati numerici interi casuali tra 1 e 99 (100 escluso).
df['NumCol2'] = np.random.randint(1, 100, size=n_rows)

# Genera una colonna di dati numerici da una distribuzione uniforme tra 0 e 1.
df['NumCol3'] = np.random.uniform(0, 1, size=n_rows)

# Calcola il numero totale di valori mancanti desiderati nel DataFrame.
total_missing_values = int(0.03 * n_rows * len(df.columns))  # 3% dei valori totali.

# Introduce valori mancanti in modo casuale nel DataFrame.
for column in df.columns:
    # Determina casualmente quanti valori mancanti inserire in ogni colonna.
    num_missing_values = np.random.randint(0, total_missing_values + 1)
    
    # Seleziona indici casuali in cui inserire i valori mancanti.
    missing_indices = np.random.choice(n_rows, size=num_missing_values, replace=False)
    
    # Imposta i valori mancanti (NaN) in quegli indici per la colonna corrente.
    df.loc[missing_indices, column] = np.nan

# Visualizza il DataFrame finale.
df
```
